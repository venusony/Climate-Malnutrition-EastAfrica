```python
# my zips live here
from pathlib import Path
import zipfile

raw = Path(r"C:\Users\venus\OneDrive\Documents\GitHub\Climate-Malnutrition-EastAfrica\data_raw\DHS_zip")

# i want all extracted files to go into data_raw\unzipped
out_root = raw.parent / "unzipped"
out_root.mkdir(parents=True, exist_ok=True)

print("zip folder:", raw)
print("unzipped folder:", out_root)

```

```python
# this will extract each zip into data_raw\unzipped\<zip_stem>
# keeping it very readable
zips = sorted(raw.glob("*.zip"))
count = 0

for zpath in zips:
    dest = out_root / zpath.stem
    dest.mkdir(parents=True, exist_ok=True)
    with zipfile.ZipFile(zpath) as zf:
        zf.extractall(dest)
    count += 1
    print("extracted:", zpath.name, "->", dest.name)

print("done. total extracted:", count)
```

```python
# i want a small table that shows each folder, how many files, and a few example names
import pandas as pd

rows = []
for folder in sorted(p for p in out_root.iterdir() if p.is_dir()):
    files = sorted(folder.glob("*"))
    rows.append({
        "folder": folder.name,
        "num_files": len(files),
        "examples": ", ".join([f.name for f in files[:5]])
    })

idx = pd.DataFrame(rows)
idx.head(15)
```

```python
docs = Path(r"C:\Users\venus\OneDrive\Documents\GitHub\Climate-Malnutrition-EastAfrica\docs")
docs.mkdir(parents=True, exist_ok=True)

out_csv = docs / "dhs_zip_index.csv"
idx.to_csv(out_csv, index=False)
print("wrote:", out_csv)
```
